```python
#%%
import pandas as pd
import requests

def get_data(year : int) -> pd.DataFrame:
    
    url_front = 'https://api.census.gov/data/'
    url_back = '/acs/acs5?get=group(B03002)&ucgid=pseudo(0400000US17$0500000)'

    url = url_front + str(year) + url_back

    response = requests.get(url)
    data = response.json()

    df_ = pd.DataFrame(data, columns=data[0]).drop(0)
    return df_

def add_titles(df_ : pd.DataFrame) -> pd.DataFrame:

    # Create a dictionary for renaming columns
    rename_dict = dict(zip(labels['code'], labels['variables']))

    # Rename the columns of df
    df_.rename(columns=rename_dict, inplace=True)

    return df_


if __name__ == '__main__':

    # -----------------------------------------------------
    # find labels
    # -----------------------------------------------------

    url = 'https://api.census.gov/data/2019/acs/acs5/groups/B03002'

    response = requests.get(url)
    data = response.json()

    labels = pd.DataFrame(data).reset_index()
    labels = labels.rename(columns={'index' : 'code'})
    labels['variables'] = labels['variables'].apply(lambda x: x['label'])

    labels.head()

    # -----------------------------------------------------
    # get data
    # -----------------------------------------------------

    # find most recent year
    for i in range(2023, 2100):

        try: get_data(i)
        except:
            final = i
            break

    # retrieve data
    data_sets = {y : get_data(y) for y in range(2010, i)}

    # -----------------------------------------------------
    # clean data
    # -----------------------------------------------------

    data_sets = {y : add_titles(df) for y, df in data_sets.items}
```
